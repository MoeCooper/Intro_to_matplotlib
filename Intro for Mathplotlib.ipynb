## Stateless way of plotting

```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y);
```

```python
# Method to plot
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y); # add some data
```

```python
# Recommended way to plot
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50,100,150,200]); # adds data
type(fig), type(ax)
```

## Matplotlib example workflow

```python
# Step 0. import matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

# Step 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# Step 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # dimensions of plot (width, height)

# Step 3. Plot the data
ax.plot(x,y)

# Step 4. Customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# Step 5. Save and show 
# fig.savefig(#insert image here)
```

```python
# make a plot from dict
```

```python
prices = {"Almonds": 10, "Peanuts": 8, "Cashews": 12}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values());
ax.set(title="Nut Prices", xlabel="Types of nuts", ylabel="Prices in $");
```

```python
student_grades = {"John": 87, "Lucy": 92, "Ashley": 88, "James": 82, "Rob": 80}
fig, ax = plt.subplots()
ax.scatter(student_grades.keys(), student_grades.values());
ax.set(Title="Student Grades for Science", xlabel="Students", ylabel="Grades");
```

## Two Options for subplots

```python
# subplot 1
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(),prices.values());
ax4.hist(np.random.randn(100));
```

```python
# Subplots option 2
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# plot to each different ixdex
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].hist(np.random.randn(100));
ax[1,1].scatter(np.random.random(10), np.random.random(10));
```
